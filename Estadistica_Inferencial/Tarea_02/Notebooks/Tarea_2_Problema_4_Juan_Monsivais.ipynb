## Función para simular Binomial Negativa

```r
moneda_BN <- function(p, N, r) #Emula una binomial negativa
{
    posiblesResultados <- c("A","S") #A: Aguila , S: Sol
    probabilidades <- c(p , 1-p) #Definimos el vector de probabilidades
    vector <- c()
    for(i in r:N)  #Número de experimentos
    {
        contador <- 0 #Numero de intentos que le tomó llegar a r exitos
        subCont <- 0 #Bandera para romper el ciclo
        while(TRUE) #Un bucle indefinido que se romperá al obtener el resultado "A"
        {
          resultado <- sample(posiblesResultados, size = 1 ,prob = probabilidades,
                      replace = T) #Tomamos un solo resultado

          contador <- contador + 1
          if(resultado == "A") # Si el resultado es Aguila, suma al subCont
          {
              subCont <- subCont + 1
              if(subCont == r) #Condición de paro
              {
                  vector <- c(vector,contador-r+1) #Guardamos el número de intentos que tomó llegar al resultado
                  break
              }

          }
        }
    }
    return (vector) #Devolvemos el vector de resultados
}
```

## Experimento 1

```r
results1 <- moneda_BN(0.2,1000000,2) #Inicio de la simulción con p = 0.2, N = 1e6, r = 2
write.csv(results1, file = "BN_1e6_0_2_2.csv") #Guardamos el CSV
```

```r
tabla1 <- table(results1)
tabla1N <- prop.table(tabla1) # Normalizamos las frecuencias para comparar con BN
```

```r
plot(tabla1N, col = "blue",
    main = "Frecuencia del intento 'x', que tomó obtener el r-ésimo éxito
    p = 0.2, r = 2, N=10^6")

BN <- dnbinom(0:max(results1), 2 ,0.2)

points(BN, col = "red")
```
