```r
moneda_Geom <- function(p, N) #Emula una geométrica
{
    posiblesResultados <- c("A","S") #A: Aguila , S: Sol
    probabilidades <- c(p , 1-p) #Definimos el vector de probabilidades
    vector <- c()
    for(i in 1:N)#Número de experimentos
    {
        contador <- 0
        while(TRUE) #Un bucle indefinido que se romperá al obtener el resultado "A"
        {
          resultado <- sample(posiblesResultados, size = 1 ,prob = probabilidades,
                      replace = T) #Tomamos un solo resultado

          contador <- contador + 1

          if(resultado == "A") # Si el resultado es Aguila, se detiene
          {
            vector <- c(vector,contador)
            break
          }
        }
    }
    return (vector) #Devolvemos el vector de resultados
}
```

## Simulación con $10^4$ pruebas

```r
probabilidades <- c(0.5,0.1,0.01)
results <- sapply(probabilidades, function(p) moneda_Geom(p, 10000)) #INICIO DE LA SIMULACIÓN
```

```r
write.csv(results, file = "datos_1e4.csv", row.names = FALSE)
```

```r
tablaF1 <- table(results[,1])   #Probabilidad = 0.5
tableF1 <- prop.table(tablaF1) #Normalizado
plot(tableF1, col = "royalblue",
     main = "Frecuencias del intento 'x' hasta obtener águila \n p=0.5, N = 1e6",
     type = "h",
     xlab = "Intento ",
     ylab = "Frecuencia de éxitos en el 'x' intento")

muestra_geom <- dgeom(0:max(results[,1]),0.5)

n <- 0:max(results[,1])

lines(n+1,muestra_geom, col = "red2", type = "o", lty = 2)

legend("topright", legend = c("Simulación","Geométrica"),
       col = c("royalblue","red2"), lty=c(1,2),lwd = c(3,1) , cex=0.9)
```
